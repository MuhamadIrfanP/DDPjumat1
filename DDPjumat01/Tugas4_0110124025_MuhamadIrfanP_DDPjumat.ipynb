```python
#menentukan bilangan ganjil dan genap
print("1.Menentukan bilangan ganjil dan genap")
#input bilangan
bilangan= int(input("masukan bilangan:"))
#proses & output
if bilangan % 2 == 0:
 print(bilangan,"merupakan bilangan genap")
else:
 print(bilangan,"merupakan bilangan ganjil")
```

```python
#Menentukan nilai ujian
print("2. Menentukan nilai ujian")
#input nilai ujian
nilai_ujian = int(input("masukan nilai ujian:"))
#proses & output
if nilai_ujian > 80:
 print ("selamat anda lulus")
else:
 print ("Remedial atau mengulang")
```

```python
#Menentukan cek usia dan status
print ("3.Menentukan usia dan status")
#input usia dan status
usia = int(input("Masukan usia"))
#proses dan input
if usia >= 18:
 print("Dewasa")
elif usia >= 13 and usia <= 17:
 print ("Remaja")
else:
 print ("anak-anak")
```
